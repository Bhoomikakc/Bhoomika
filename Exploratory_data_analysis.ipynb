### When you start exploring your data, always start by exploring the main feature of your data. For example, as we are working on a dataset based on Instagram Reach, we should start by exploring the feature that contains data about reach. In our data, the Impressions column contains the data about the reach of an Instagram post. So let’s have a look at the distribution of the Impressions:

```python
df["Impressions"].plot.hist()
```

### the number of impressions on each post over time:

```python
plt.plot(df.index,"Impressions",data=df)
```

### Now let’s have a look at all the metrics like Likes, Saves, and Follows from each post over time:

```python
plt.figure(figsize=(8,8))
plt.plot(df.index,"Likes",data=df)
plt.plot(df.index,"Saves",data=df)
plt.plot(df.index,"Follows",data=df)
plt.legend()
plt.show()
```

### Now let’s have a look at the distribution of reach from different sources

```python
reach_sources = ['From Home', 'From Hashtags', 'From Explore', 'From Other']
reach_counts = [df[source].sum() for source in reach_sources]
```

```python
reach_sources
```

```python
reach_counts
```

```python
plt.pie(reach_counts,labels=reach_sources,autopct="%0.1f%%")
```

### Now let’s have a look at the distribution of engagement sources:

```python
engagement_source=["Likes","Comments","Saves","Shares"]
engagement_counts=[df[source].sum() for source in engagement_source]
```

```python
plt.pie(engagement_counts,labels=engagement_source,autopct="%0.1f%%")
plt.show()
```

### Now let’s have a look at the relationship between the number of profile visits and follows

```python
plt.scatter(x="Profile Visits",y="Follows",data=df)
```

```python
sns.heatmap(df.corr())
```

### Exploratory data analysis (EDA) is a Data Science concept where we analyze a dataset to discover patterns, trends, and relationships within the data. It helps us better understand the information contained in the dataset and guides us in making informed decisions and formulating strategies to solve real business problems
